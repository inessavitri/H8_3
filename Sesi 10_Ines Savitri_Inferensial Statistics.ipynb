### Confidence Intervals

```python
# CI for the population Proportion in Python
import pandas as pd
import numpy as np
df = pd.read_csv('heart.csv')
df
```

```python
df['Sex1'] = df.sex.replace({1: "Male", 0: "Female"})
```

```python
dx = df[["target", "Sex1"]].dropna()
```

```python
pd.crosstab(dx.target, dx.Sex1)
```

```python
p_fm = 226/(86+226)
p_fm
```

```python
n = 86+226
n
```

```python
se_female = np.sqrt(p_fm * (1 - p_fm) / n)
se_female
```

```python
z_score = 1.96
lcb = p_fm - (z_score * se_female) # lower limit of the CI
ucb = p_fm + (z_score * se_female) # upper limit
lcb, ucb
```

```python
# menghitung menggunakan library 'statsmodels'
import statsmodels.api as sm
sm.stats.proportion_confint(n * p_fm, n)
```

### Critical Z-Scores

```python
tail = 1 - stats.norm.cdf(1.96)
tail
```

```python
# The area under the curve between the z-critical scores is approximately 0.95
1 - (tail * 2)
```

```python
lower_95_perc_bound = mean_of_sample_means - 1.96 * standard_error
lower_95_perc_bound 
```

## Hypothesis Testing for One Proportion

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import scipy.stats.distributions as dist
df = pd.read_csv('heart.csv')
df.head()
```

- H0: p0 = 0.42  # null hypothesis
- Ha: p > 0.42  # alternative hypothesis

```python
p_us = len(df[df['target']=='1'])/len(df)
p_us
```

```python
se = np.sqrt(0.42 * (1-0.42) / len(df))
se
```

```python
# Best estimate
be = p_us

# hypothesized estimate
he = 0.42
test_stat = (be - he)/se
```

```python
test_stat
```

```python
pvalue = 2*dist.norm.cdf(-np.abs(test_stat))
pvalue
```

### Statisrical Hypothesis Test Cheat sheet (optional)

#### Normality Tests
This section lists statistical tests that you can use to check if your data has a Gaussian Distribution


Assumptions:
- Observations in each sample are independent and identically distributed (iid)

Interpretation:
- H0: the sample has a Gaussian distribution
- H1: the sample does not have a Gaussian distribution


#### Shapiro-Wilk Test

```python
# Example of the Shapiro-Wilk Normality Test

from scipy.stats import shapiro
data = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]
stat, p = shapiro(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably Gaussian')
else:
    print('Probably not Gaussian')
```

#### D'Agostino's K^2 Test

```python
# Example of the D'Agostino's K^2 Normality Test

from scipy.stats import normaltest
data = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]
stat, p = normaltest(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably Gaussian')
else:
    print('Probably not Gaussian')
```

#### Anderson-Darling Test

```python
# Example of the Anderson-Darling Normality Test

from scipy.stats import anderson
data = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]
result = anderson(data)
print('stat=%.3f' % (result.statistic))
for i in range(len(result.critical_values)):
      sl, cv = result.significance_level[i], result.critical_values[i]
      if result.statistic < cv:
          print('Probably Gaussian at the %.1f%% level' % (sl))
      else:
          print('Probably not Gaussian at the %.1f%% level' % (sl))
```

### Correlation Tests

#### Pearson's Correlation Coefficient

```python
# Example of the Pearson's Correlation test
from scipy.stats import pearsonr
data1 = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]
data2 = [0.353, 3.517, 0.125, -7.545, -0.555, -1.536, 3.350, -1.578, -3.537, -1.579]
stat, p = pearsonr(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably independent')
else:
    print('Probably dependent')
```

### Spearman's Rank Correlation

```python
# Example of the Spearman's Rank Correlation test
from scipy.stats import spearmanr
data1 = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]
data2 = [0.353, 3.517, 0.125, -7.545, -0.555, -1.536, 3.350, -1.578, -3.537, -1.579]
stat, p = spearmanr(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably independent')
else:
    print('Probably dependent')
```

### Kendall's Rank Correlation

```python
# Example of the Kendall's Rank Correlation test
from scipy.stats import kendalltau
data1 = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]
data2 = [0.353, 3.517, 0.125, -7.545, -0.555, -1.536, 3.350, -1.578, -3.537, -1.579]
stat, p = kendalltau(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably independent')
else:
    print('Probably dependent')
```

### Chi-Squared Test

```python
# Example of Chi-Squared test
from scipy.stats import chi2_contingency
table = [[10, 20, 30],[6, 9, 17]]
stat, p, dof, expected = chi2_contingency(table)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably independent')
else:
    print('Probably dependent')
```
